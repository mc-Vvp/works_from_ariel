```python
#question 1

my_movies = [['How I Met Your Mother', 'Friends', 'Silicon Valley'],
             ['Family Guy', 'South Park', 'Rick and Morty'],
             ['Breaking Bad', 'Game of Thrones', 'The Wire']]

count = 0
for i in range(len(my_movies)):
    for j in range(len(my_movies[i])):
        count = len(my_movies[i][j])
        print("The title '{}' is '{}' characters long.".format(my_movies[i][j], count))
```

```python
#question 2

i = 0
print(i)
i += 1
while i<=99:
    if i%3 == 0 and i%5 != 0:
        print("Boom!")
    elif i%3 != 0 and i%5 == 0:
        print("Bim!")
    elif i%3 == 0 and i%5 == 0:
        print("BimBamBom!")
    else:
        print(i)
    i += 1
```
